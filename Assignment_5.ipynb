```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as st
```

```python
data = pd.read_excel("Housing.xlsx",index_col=0)
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
for i in data.columns:
    if (i !="PropertyAddress" and i != 'LandUse' and i != 'SaleDate' and i != 'SalePrice'):
        data[i].fillna(st.mode(data[i]),inplace=True)
```

```python
data
```

```python
data.rename(columns={"PropertyAddress":"Location"},inplace=True)
data.rename(columns={"SalePrice":"Cost"},inplace=True)
```

```python
data
```

```python
data['year'] = pd.DatetimeIndex(data['SaleDate']).year
data['month'] = pd.DatetimeIndex(data['SaleDate']).month
```

```python
data
```

```python
data['pin']=data['Location'].str[:4]
data
```

**1. How has the price of properties changed over the years**

```python
c=data.groupby("year")["Cost"].count()
sns.barplot(x=c.index,y=c.values)
```

**2. Does a specific month or span of months have the highest property prices?**


```python
data.groupby("year")["Cost"].sum()
```

```python
c=data.groupby("month")["Cost"].count()
sns.barplot(x=c.index,y=c.values)
```

**3. For what purpose were the properties generally purchased?**

```python
p=data.groupby("LandUse")["LandUse"].count()
p
```

```python
sns.barplot(x=p.values,y=p.index)
```

**4. What is the average age of the properties?**



```python
data['Age']=2022-data['YearBuilt']
data
```

```python
avg=np.mean(data.Age)
avg
```

**5. Have the age of houses affected the price?**

```python
d=data.groupby("Age")["Cost"].count()
d
sns.scatterplot(x=d.index,y=d.values)
```

**6. Have the number of rooms in the house affected the price?**



```python
data['totalrooms']=data['Bedrooms']+data['FullBath']+data['HalfBath']
data
```

```python
j=data.groupby("totalrooms")["Cost"].count()
sns.scatterplot(x=j.index, y=j.values)
```

```python
sns.barplot(x=data["totalrooms"],y=data["Cost"])
```

**7.Does the size of the property (no. of acres) have any influence in its price?**

```python
g=data.groupby("Acre")["Cost"].count()
sns.scatterplot(x=g.index, y=g.values)
```

**8.Are houses located in specific pin numbers expensive over the others?**

```python
g=data.groupby(data["pin"])["Cost"].sum()
g
```

```python
sns.scatterplot(x=g.values,y=g.index)
```
